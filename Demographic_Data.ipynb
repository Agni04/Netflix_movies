**Importing required libraries:**

```python
import pandas as pd
import numpy as np
```

```python
#importing data:
df = pd.read_csv("adult.data.csv")
df
```

How many of each race are represented in this dataset? This should be a Pandas series with race names as the index labels.

```python
race_count =df['race'].value_counts()
race_count
```

What is the average age of men?

```python
average_age_men = round(df.loc[df["sex"]=="Male"]["age"].mean(),1)
average_age_men
```

What is the percentage of people who have a Bachelor's degree?

```python
num_bachelors = len(df.loc[df["education"]=="Bachelors"])
total=len(df["education"])
percentage_bachelors = round((num_bachelors / total )*100,1)
percentage_bachelors
```

What percentage of people with advanced education (Bachelors, Masters, or Doctorate) make more than 50K? What percentage of people without advanced education make more than 50K? with_ and without Bachelors, Masters, or Doctorate

```python
higher_education =df.loc[df["education"].isin(['Bachelors','Masters','Doctorate'])]
lower_education = df.loc[~df["education"].isin(['Bachelors','Masters','Doctorate'])]
```

percentage with salary >50K

```python
non_percentage_higher=len(higher_education[higher_education.salary=='>50K'])
higher_education_rich = round(non_percentage_higher/len(higher_education)*100,1)
higher_education_rich
```

```python
non_percentage_lower=len(lower_education[lower_education.salary=='>50K'])
lower_education_rich =round(non_percentage_lower/len(lower_education)*100,1)
lower_education_rich
```

What is the minimum number of hours a person works per week (hours-per-week feature)?

```python
min_work_hours = df['hours-per-week'].min()
min_work_hours
```

What percentage of the people who work the minimum number of hours per week have a salary of >50K

```python
num_min_workers = df.loc[df['hours-per-week']==min_work_hours]
num_min_workers
```

```python
rich_percentage =round(len(num_min_workers[num_min_workers.salary=='>50K'])/len(num_min_workers)*100,1)
rich_percentage
```

 What country has the highest percentage of people that earn >50K?

```python
country_count=df['native-country'].value_counts()
country_rich_country=df.loc[df['salary']=='>50K']['native-country'].value_counts()
highest_earning_country =(country_rich_country/country_count*100).idxmax()
highest_earning_country_percentage =round(country_rich_country/country_count*100,1).max()
highest_earning_country
```

```python
highest_earning_country_percentage
```

 Identify the most popular occupation for those who earn >50K in India.

```python
people_of_india_rich=df.loc[(df['native-country']=='India')&(df["salary"]=='>50K')]
people_of_india_rich
```

```python
occupation_counts=people_of_india_rich['occupation'].value_counts()
top_IN_occupation =occupation_counts.idxmax()
top_IN_occupation
```
